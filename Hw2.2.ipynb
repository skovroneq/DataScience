```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("2017_jun_final - 2017_jun_final.csv")
```

```python
spaces_proportion = data.isnull().sum() / len(data)
print(f"Spaces proportion:\n{spaces_proportion}")
```

```python
column_to_keep =["Programming language"]
columns_to_drop = [col for col in data.columns if (col not in column_to_keep and data[col].isnull().any())]
new_data = data.drop(columns=columns_to_drop)
new_data.head()
```

```python
spaces_proportion_cleaned = new_data.isnull().sum() / len(new_data)
print(f"Spaces proportion:\n{spaces_proportion_cleaned}")
```

```python
data = data.dropna()
data.head()
```

```python
new_data.shape
```

```python
python_data = new_data[new_data["Programming language"] == "Python"]
python_data.head()
```

```python
python_data.shape
```

```python
python_grouped = python_data.groupby(by=['Role'])
```

```python
python_aggregated = python_grouped.agg({'Salary/month' : ['min', 'max']})
python_aggregated.head()
```

```python
def fill_avg_salary(n):
    avg = sum(n)/len(n)
    return avg
```

```python
avg_salary = python_data.groupby("Role")["Salary/month"].apply(fill_avg_salary)
python_aggregated["Avg salary/month"] = list(avg_salary)
python_aggregated.head()
```

```python
stats = python_aggregated["Avg salary/month"]
stats.describe()
```

```python
python_aggregated.to_csv('python_aggregated.csv')
```

```python
plt.plot(python_aggregated.index, python_aggregated['Salary/month', 'min'], label='Min Salary', color='orange', linestyle=":",marker="D", markersize=8)
plt.plot(python_aggregated.index, python_aggregated['Salary/month', 'max'], label='Max Salary', color='brown', linestyle="--",marker="o", markersize=8)
plt.xticks(rotation=45)
plt.ylabel('Salary/month', fontsize="small")
plt.title('Python programmers salary depending on role', fontsize="small")
plt.grid(visible=True)
plt.legend()
```

```python
labels = python_data["Role"].unique()
roles = python_data["Role"].value_counts()

data = [roles[label] for label in labels]
explode = list(np.ones(len(data))*0.08)

plt.pie(
    x=data,
    labels=labels,
    shadow=True,
    autopct="%.2f%%",
    explode=explode,
)
plt.title("Percentage of roles")
```
