__Zadanie 1__

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data_2d = pd.read_csv("data_2d.csv")
mnist = pd.read_csv("mnist.csv")

inertia_2d = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_2d)
    inertia_2d.append(kmeans.inertia_)

plt.plot(range(1, 15), inertia_2d, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for data_2d')
plt.show()

inertia_mnist = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(mnist)
    inertia_mnist.append(kmeans.inertia_)

plt.plot(range(1, 15), inertia_mnist, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for mnist')
plt.show()
```

Na podstawie wykresów można stwierdzić, że optymalna liczba klastrów dla zbioru danych 2D wynosi 2, a dla zbioru danych mnist 3.

__Zadanie 2__

```python
from sklearn.decomposition import PCA

def visualize_clustering(data, optimal_clusters, title):
    kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
    kmeans.fit(data)
    
    plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', s=50, alpha=0.5)
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.colorbar(label='Cluster')
    plt.show()

# 2D dataset
optimal_clusters_2d = 2
visualize_clustering(data_2d.values, optimal_clusters_2d, 'Clustering result for 2D dataset')

# MNIST after PCA
optimal_clusters_mnist = 3
pca = PCA(n_components=2, random_state=42)
mnist_pca = pca.fit_transform(mnist)
visualize_clustering(mnist_pca, optimal_clusters_mnist, 'Clustering result for MNIST after PCA')
```
